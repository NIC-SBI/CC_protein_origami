```python
import collections
def permute_segment_left(segs, n=1):
    """Permutes a segment to the left for n movements"""
    q = collections.deque(segs)
    q.rotate(-n)
    return list(q)

assert permute_segment_left(['a', 'b', 'c']) == \
                            ['b', 'c', 'a']
```

```python
import collections
def permute_vertices_left(verts, n=1):
    """Permutes a segment to the left for n movements"""
    q = collections.deque(verts)
    #last vertex is same as first
    last = q.pop()
    q.rotate(-n)
    q.append(q[0])
    return list(q)

assert permute_vertices_left([1,2,3,1], 1) == [2, 3, 1, 2]
assert permute_vertices_left([1,2,3,1], 2) == [3, 1, 2, 3]
                                                        
```

```python
def get_permutation_name(basename, permutation=1):
    """gets the name for the permutation with a, b, c...
    Currently does not work for negative"""
    return basename+chr(permutation+ord('a'))
    
assert get_permutation_name('p1',0) == "p1a"
assert get_permutation_name('p1',1) == "p1b"
```

```python
def get_segment_distances(segment_topology):
    """given an array of char as segments find the distance between pairs in the sequnce"""
    st = [c.lower() for c in segment_topology]
    #delete duplicates
    blocks = sorted(list(set(st)))
    
    dist = []
    for block in blocks:
    
        inds = [i for i, x in enumerate(st) if x == block]
        assert len(inds) == 2, block+ " does not have a pair: "+ str(len(inds))
        dist.append(abs(inds[1]-inds[0]-1))
    return dist
```

```python
import numpy as np
import pandas as pd
   
segments = [u.vertex_to_segmet(p) for p in permutations]

for s in segments:
    print("-".join(s))    
    
for name, s in zip(names,segments):
    dist = get_segment_distances(s)
    print(name, "min:", np.min(dist),
                "max:", np.max(dist),
                "avg:", np.mean(dist),
                "std:", np.std(dist))
    
```

```python
#generate empty table
allpnames = []
for n,v in enumerate(four_pyramid_data):
    s = u.vertex_to_segmet(v)
    for p in range(len(s)):
        allpnames.append(get_permutation_name('p'+str(n+1),p))
                         
len(allpnames)
```

```python
df = pd.DataFrame(index=allpnames,columns="min max avg stdev segs vert".split())
for n,v in enumerate(four_pyramid_data):
    s = u.vertex_to_segmet(v)
    for p in range(len(s)):
        name = get_permutation_name('p'+str(n+1),p)
        #distances
        ps = permute_segment_left(s, p)
        pv = permute_vertices_left(v, p)
        dist = get_segment_distances(ps)
        df.loc[name, 'min'] = np.min(dist)
        df.loc[name, 'max'] = np.max(dist)
        df.loc[name, 'avg'] = np.mean(dist)
        df.loc[name, 'stdev'] = np.std(dist)
        df.loc[name, 'segs'] = "-".join(ps)
        df.loc[name, 'vert'] = str(pv)
```

```python
df=df.sort_values(by='avg stdev min'.split())
```
